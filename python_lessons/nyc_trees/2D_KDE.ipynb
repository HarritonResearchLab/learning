```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/HarritonResearchLab/learning/main/python_lessons/nyc_trees/nyc_1995_trees.csv')

df = df.drop(labels=['LatinSpecies', 'Borough'], axis=1)

raw_longs = np.array(df['Longitude'])
raw_lats = np.array(df['Latitude'])

longs = np.array([])
lats = np.array([])

# clean bad data
for i, j in zip(raw_longs, raw_lats):
    if i!=1.0 and j!=1.0:
        longs = np.append(longs, i)
        lats = np.append(lats, j)

# Plot
plt.style.use('fivethirtyeight')
plt.rcParams['font.family']='serif'
ax = sns.kdeplot(longs, lats, cmap='Blues', shade=True)
plt.xlabel('Longitude', fontsize=15)
plt.ylabel('Latitude', fontsize=15)
plt.title('NYC Tree Density Map', fontsize=17)
```
